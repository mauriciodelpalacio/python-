### recursive function to get a factorial number, a recursive function its a function  that defines by itself. its a way to get compact yet powerful functions 

```python
def factorial(x):
    ## Recursive way of doing a factorial number
    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))


num = 53
print("The factorial of", num, "is", factorial(num))
```
